# Contadores sin Locks

En este experimento se analiza el comportamiento de un contador compartido 
incrementado por múltiples hilos **sin utilizar mecanismos de sincronización**.
 
El objetivo es observar los **efectos de las condiciones de carrera (race conditions)** 
sobre el resultado final y el desempeño del programa.

## Contexto

Cada hilo ejecuta un número fijo de incrementos sobre una variable compartida `value`.
 
Cuando se ejecuta sin mecanismos de exclusión mutua (como `mutex`), múltiples hilos pueden leer y escribir el valor del contador al mismo tiempo, produciendo resultados inconsistentes.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Datos experimentales obtenidos
data = {
    "Prueba": [1, 2, 3, 4, 5],
    "Número de hilos": [1, 2, 4, 8, 16],
    "Valor esperado": [1_000_000, 2_000_000, 4_000_000, 8_000_000, 16_000_000],
    "Valor obtenido": [1_000_000, 1_030_174, 1_039_790, 1_309_698, 1_177_806],
    "Tiempo (seg)": [0.004213, 0.004213, 0.025465, 0.044530, 0.088573]
}

df = pd.DataFrame(data)
df
```

## 1. Gráfica de desempeño

A continuación se muestra el **tiempo de ejecución** del programa en función del número de hilos. En principio, podríamos esperar que más hilos reduzcan el tiempo de ejecución. Sin embargo, al no haber sincronización, el comportamiento puede ser irregular debido a la interferencia entre hilos y la sobrecarga de cambio de contexto.

```python
plt.figure(figsize=(8,5))
plt.plot(df["Número de hilos"], df["Tiempo (seg)"], marker='o', linestyle='-', linewidth=2)
plt.title("Desempeño del contador sin locks")
plt.xlabel("Número de hilos")
plt.ylabel("Tiempo de ejecución (segundos)")
plt.grid(True)
plt.show()
```

## 2. Comparación del valor esperado vs. valor obtenido

En esta gráfica se compara el valor **esperado** (sin pérdida de operaciones) con el valor **realmente obtenido** en la ejecución.

La diferencia se debe a las **condiciones de carrera**, ya que varias operaciones `value++` se ejecutan de forma concurrente sin exclusión mutua, sobrescribiendo los resultados parciales.

```python
plt.figure(figsize=(8,5))
plt.plot(df["Número de hilos"], df["Valor esperado"], marker='o', label="Valor esperado")
plt.plot(df["Número de hilos"], df["Valor obtenido"], marker='s', label="Valor obtenido")
plt.title("Comparación del valor esperado vs obtenido")
plt.xlabel("Número de hilos")
plt.ylabel("Valor del contador")
plt.legend()
plt.grid(True)
plt.show()
```

## 3. Porcentaje de error

A continuación se calcula el **porcentaje de error** del valor obtenido con respecto al valor esperado. Este error aumenta con el número de hilos,evidenciando la pérdida de actualizaciones por interferencia concurrente.

```python
df["Error (%)"] = (1 - df["Valor obtenido"] / df["Valor esperado"]) * 100

plt.figure(figsize=(8,5))
plt.bar(df["Número de hilos"], df["Error (%)"], color="tomato")
plt.title("Porcentaje de error respecto al valor esperado")
plt.xlabel("Número de hilos")
plt.ylabel("Error (%)")
plt.grid(axis='y')
plt.show()

df
```

## Conclusiones

Al ejecutar el experimento podemos ver que:
- El experimento demuestra que las operaciones sobre variables compartidas no son **atómicas**.
- Sin mecanismos de sincronización (como `mutex` o variables atómicas), los resultados se vuelven **inconsistentes**.
- A medida que se incrementa el número de hilos:
  - Aumentan las colisiones y pérdidas de incrementos.
  - El desempeño no mejora e incluso puede degradarse.
 
> Es posible puede repetir este experimento implementando un **contador con locks (`pthread_mutex_t`)** o con **operaciones atómicas**, para comparar el impacto en la **exactitud y el rendimiento**.
